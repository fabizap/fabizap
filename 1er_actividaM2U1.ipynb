### Media muestral del importe mensual cobrado


```python
media_Impormensual = df['Impormensual'].mean() # Calcula la media de la columna.
media_Impormensual # Muestra el valor de la media calculada.
```

### Varianza y desvío estándar muestral del importe mensual cobrado

```python
varianza_Impormensual = df['Impormensual'].var(ddof=1)  # Calcula la varianza de la columna. El parámetro ddof=1 especifica que se está utilizando el estimador insesgado (n-1).
varianza_Impormensual  # Muestra el valor de la varianza calculada.
```

```python
desvio_Impormensual = df['Impormensual'].std(ddof=1)  # Calcula la desviación estándar de la columna. El parámetro ddof=1 especifica que se está utilizando el estimador insesgado (n-1).
desvio_Impormensual  # Muestra el valor de la desviación estándar calculada.
```
